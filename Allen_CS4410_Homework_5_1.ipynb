Load the Dataset

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
from sklearn.cluster import KMeans
```

Set up range for k and array to hold inertias

```python
K = range(1, 11)
inertias = []
```

Check each value of k in range and save inertias

```python
for k in K:
  kmeans = KMeans(n_clusters=k, random_state=11)
  kmeans.fit(iris.data)
  inertias.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

Graph the inertia for each value of k

```python
plt.plot(K, inertias)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('The Elbow Method Using Inertia')
plt.show()
```
