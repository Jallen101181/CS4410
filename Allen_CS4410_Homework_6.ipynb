Import Hamlet from Gutenberg

```python
import requests
```

```python
target_url ='http://www.gutenberg.org/files/2265/2265.txt'

response = requests.get(target_url)

data = response.text
```

Create a TextBlob and import stopwords

```python
from textblob import TextBlob
```

```python
blob = TextBlob(data)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

Create an items list and remove stopwords

```python
items = blob.word_counts.items()
```

```python
items = [item for item in items if item[0] not in stop_words]
```

Sort items by frequency

```python
from operator import itemgetter
```

```python
sorted_items = sorted(items, key=itemgetter(1), reverse=True)
```

```python
top20 = sorted_items[0:20]
```

Import to a dataframe and graph

```python
import pandas as pd
```

```python
 df = pd.DataFrame(top20, columns=['word', 'count'])
```

```python
df
```

```python
axes = df.plot.bar(x='word', y='count', legend=False)
```

```python
from wordcloud import WordCloud
```

Create wordcloud using Hamlet and provided mask image

```python
wordcloud = WordCloud(colormap='prism', mask=mask_image, background_color='white', stopwords=stop_words)
```

```python
wordcloud = wordcloud.generate(data)
```
