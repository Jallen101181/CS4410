```python
#Import dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#Import KNN model and kfold, cross value scoring
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, cross_val_score

knn = KNeighborsClassifier()
```

```python
#Loop for all values of k between 1 and 31
import matplotlib.pyplot as plt

k_range = range(1, 31)

k_plot = []
acc_plot = []

X = iris.data
y = iris.target

for k in k_range:
  knn = KNeighborsClassifier(n_neighbors=k)
  scores = cross_val_score(knn, X, y, cv=10)
  k_plot.append(k)
  acc_plot.append(scores.mean())
```

```python
#Plot the mean accuracy of each value of k
plt.plot(k_plot, acc_plot)

plt.xlabel("Value of k for KNN")
plt.ylabel("Cross-Validated Accuracy")

plt.title("Cross-Validated Accuracy vs. Value of k for KNN")

plt.show()
```

```python
#Import gridsearchcv
from sklearn.model_selection import GridSearchCV
```

```python
#Create the needed dictionary
k_range = list(range(1, 31))
param_grid = dict(n_neighbors=k_range)
```

```python
#Use GridSearchCV to find the best value for k
grid = GridSearchCV(knn, param_grid, cv=10, scoring='accuracy', return_train_score=False)
grid.fit(X, y)
print(grid.best_score_)
print(grid.best_params_)
```

Both the plot from the loop and the gridsearchcv show the k=13 is the best value to choose for the KNN
